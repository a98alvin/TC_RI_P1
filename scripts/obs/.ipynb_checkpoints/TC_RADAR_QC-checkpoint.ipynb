```python
# Open file list function
def open_list(f_name):
    import pickle
    with open(f_name, "rb") as fp:   # Unpickling
        b = pickle.load(fp)
        return b
    
def save_list(f_name,list_des):
    import pickle
    with open(f_name, "wb") as fp:   #Pickling
        pickle.dump(list_des, fp)
    
from PIL import Image
from IPython.display import display
```

```python
# Write code for QC-Check
all_image_paths = open_list('/Users/acheung/data/pickles/QC_TC_RADARImages_list.pkl')

# This allows only a list to made if we want a new entry! So we can come back and keep going!

if 'good_case_loop_list' in locals():
    pass
else:
    good_case_loop_list = []

for image_ind in range(len(good_case_loop_list),len(all_image_paths)):
    print(image_ind)
    image_path = all_image_paths[image_ind][1]
    img = Image.open(image_path)
    display(img)
    
    # Ask user whether image is okay
    
    QC_boo = int(input('Pass? (y: 1, n:0): '))
    
    # Record information about which image is okay
    
    if QC_boo == 1:
        good_case_loop = all_image_paths[image_ind][0]
        good_case_loop_list.append((good_case_loop,image_path))
    else:
        continue
        


```

```python
save_list('/Users/acheung/data/pickles/QC_results',good_case_loop_list)
```
